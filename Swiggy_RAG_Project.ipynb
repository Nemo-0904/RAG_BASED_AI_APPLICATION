```python
from langchain_community.document_loaders import PyPDFLoader

pdf_path = "/content/swiggy_report (1).pdf"

loader = PyPDFLoader(pdf_path)
documents = loader.load()

print("Total pages:", len(documents))
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=600,
    chunk_overlap=100
)

chunks = text_splitter.split_documents(documents)

print("Total chunks:", len(chunks))
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

print("Embedding model loaded")
```

```python
from langchain_community.vectorstores import FAISS

vector_db = FAISS.from_documents(chunks, embedding_model)

print("Vector database created")
```

```python
query = "What is Swiggy's business model?"

results = vector_db.similarity_search(query, k=3)

for i, doc in enumerate(results):
    print("\nResult", i+1)
    print(doc.page_content[:400])
```

```python
import google.generativeai as genai
import os
from dotenv import load_dotenv

load_dotenv()
api_key = os.getenv("GEMINI_API_KEY")
if not api_key:
    raise ValueError("GEMINI_API_KEY not found in environment variables. Please set it in .env file")
genai.configure(api_key=api_key)
```

```python
model = genai.GenerativeModel("models/gemini-2.5-flash")
```

```python
import google.generativeai as genai

for m in genai.list_models():
    if "generateContent" in m.supported_generation_methods:
        print(m.name)
```

```python
while True:

    query = input("\nAsk a question about the Swiggy report (type 'exit' to stop): ")

    if query.lower() == "exit":
        print("Session ended.")
        break

    results = vector_db.similarity_search(query, k=7)

    context = "\n\n".join([doc.page_content for doc in results])

    prompt = f"""
    You are answering questions based only on the provided context from a company annual report.

    If the answer is not present in the context, reply with:
    "Not found in the provided document."

    Context:
    {context}

    Question:
    {query}

    Answer:
    """

    response = model.generate_content(prompt)

    print("\nAnswer:")
    print(response.text)
```
